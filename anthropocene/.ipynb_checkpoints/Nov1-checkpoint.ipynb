```python
import anthropocene.util as util

def overlap(listA, listB):
    """
    Return number of shared objects between listA, listB as a list
    """
    if (listA is None) or (listB is None):
        return []
    else:
        return list(set(listA) & set(listB))
    
name_list_i= ['RUGGIERO G', 'TRIARICO L', 'TROMBATORE', 'BATTISTA', "DELL'ACQUA", 'RIZZARI', 'RICCARDI L']
name_list_j= ['RUGGIERO G', 'TRIARICO L', 'TROMBATORE', 'BATTISTA', "DELL'ACQUA", 'RIZZARI', 'RICCARDI G']
overlap = util.overlap(name_list_i, name_list_j)    

print overlap
    
```
